```python
from datasets import print_shape, sort_dataset, make_polynomial
from plots import plot2D, plot3D
```

```python
X, y = make_polynomial(n_samples=1000, n_features=1, degree=2, noise=0.0, random_state=541)
X, y = sort_dataset(X, y)
```

```python
print_shape(X, y)
```

```python
title = "2D Synthetic Polynomial Regression Dataset"
xlabel = "Feature 1"
ylabel = "Target 1"
plot2D(title, xlabel, ylabel, X, y)
```
